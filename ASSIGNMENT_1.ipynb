**1. Explain the role of weights in a neuron**

Weights in a neuron determine the importance of each input feature. Every input is multiplied by a corresponding weight, which controls how much influence that input has on the final output. Larger weights indicate stronger influence, while smaller or negative weights reduce or reverse the impact. During training, these weights are updated using learning algorithms like backpropagation to minimize error. In simple terms, weights help the neural network learn patterns from data and make accurate predictions.

**2. What is an activation function?**

An activation function is a mathematical function applied to the output of a neuron to determine its final value. It introduces non-linearity into the model, allowing neural networks to learn complex patterns. Without activation functions, the network would behave like a simple linear model and fail to solve real-world problems. Common activation functions include ReLU, Sigmoid, and Tanh. For example, the Sigmoid function converts outputs into values between 0 and 1, making it useful for binary classification tasks.

**3. Define probability distribution in ML context**

In machine learning, a probability distribution describes how the values of a variable are distributed and the likelihood of each possible outcome. It helps in understanding the behavior and uncertainty of data. For example, a normal distribution shows that most values are concentrated around the mean. Many machine learning algorithms rely on probability distributions to make predictions and decisions. They are especially important in models like Naive Bayes and generative models, where probabilities are used to estimate outcomes.

**4. What is gradient in optimization?**

The gradient is a vector that represents the direction and rate of the steepest increase of a function. In optimization, particularly in gradient descent, it is used to minimize the loss function by updating model parameters such as weights. The parameters are adjusted in the opposite direction of the gradient to reduce error. A larger gradient results in bigger updates, while a smaller gradient leads to slower learning. It is a fundamental concept in training neural networks and improving model performance.

## **5. Create a project using logistic regression with TensorFlow (Colab Code)**

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

X = np.array([[25], [30], [35], [40], [45], [50]])
y = np.array([0, 0, 0, 1, 1, 1])

model = keras.Sequential([
    keras.layers.Dense(1, activation='sigmoid', input_shape=(1,))
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X, y, epochs=100)

test = np.array([[33]])
prediction = model.predict(test)

print(prediction)
```

```python
import matplotlib.pyplot as plt

history = model.fit(X, y, epochs=100, verbose=0)

plt.plot(history.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epochs')
plt.show()

plt.plot(history.history['accuracy'])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Epochs')
plt.show()
```
